```python
import pyecharts
from pyecharts.charts import Map
import pandas as pd
import matplotlib.pyplot as plt
from pyecharts import options as opts
from pyecharts.globals import ThemeType
```

```python
data = pd.read_excel('水资源.xlsx')
data2018 = data.loc[1:31]
type(data2018)
rl2018 = data2018['region'].tolist()
pl2018 = data2018['precipitation'].tolist()
wr2018 = data2018['total_water_resource'].tolist()
wc2018 = data2018['total_water_consumption'].tolist()
rs2018 = data2018['reservoir_storage'].tolist()
ac2018 = data2018['agricultural_consumption'].tolist()
```

```python
list_2018_agricultural_consumption = [[rl2018[i],ac2018[i]] for i in range(len(rl2018))]

c = (Map(init_opts=opts.InitOpts(width="1000px", height="600px",theme=ThemeType.ROMANTIC)) #可切换主题
    
     .set_global_opts(
        title_opts=opts.TitleOpts(title="2018 Agricultural Consumption Distribution map "),
         
        visualmap_opts=opts.VisualMapOpts(
            min_=100,
            max_=2000,
            range_text = ['Agricultural usage: (unit 10^8 m^3)', ''],  #分区间
            is_piecewise=True,  #定义图例为分段型，默认为连续的图例
            type_="color",
            pos_top= "bottom",  #分段位置
            pos_left="left",
            orient="vertical",
            split_number=6,  #分成6个区间
            pieces=[{"min":0,"max":50,"label":"0-50","color":"#FFF0F5"},
                    {"min":51,"max":100,"label":"51-100","color":"#FFEBCD"},
                    {"min":101,"max":200,"label":"101-200","color":"#FFDAB9"},
                    {"min":201,"max":300,"label":"201-300","color":"#F08080"},
                    {"min":301,"max":400,"label":"301-400","color":"#CD5C5C"},
                    {"min":401,"label":"more than 400","color":"#8B0000"}
                   ]         
    )
    )    
    .add("Agricultural Consumption Distribution",list_2018_agricultural_consumption,maptype="china")
    .render("2018 Agricultural Consumption Distribution.html")
)
```

```python
import seaborn as sns
from matplotlib.font_manager import FontProperties
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

data = pd.read_excel('WaterResource.xlsx')
data2018 = data.loc[1:31]
data2017 = data.loc[33:63]
data2016 = data.loc[65:95]
f, ax= plt.subplots(figsize = (12, 8))
sns.barplot(x=data2018['region'],y=data2018['reservoir_numbers'],ax=ax)
ax.set_title('2016 Reservoir Storage', fontsize=18)
sns.set_style("white") 

```

```python
f, ax= plt.subplots(figsize = (10,6))
sns.barplot(x=data2018['region'],y=data2018['reservoir_storage'],ax=ax)
ax.set_title('2018 Reservoir Storage', fontsize=18)
sns.set_style("white") 
```

```python
data2018_store = data2018.sort_values(by=['reservoir_storage'],ascending=False).head()
data2018_store
```

```python
data2018_csump = (data2018.sort_values(by=['total_water_consumption'],ascending=False)).head()
data2017_csump = (data2017.sort_values(by=['total_water_consumption'],ascending=False)).head()
data2016_csump = (data2016.sort_values(by=['total_water_consumption'],ascending=False)).head()
```

```python
plt.figure(figsize=(12,4), dpi=80)
plt.figure(1)
ax1 = plt.subplot(131)
sns.barplot(x=data2016_csump['region'],y=data2016_csump['total_water_consumption'],palette=sns.color_palette("Blues_r"))
ax1.set_title('2016 Reservoir Numbers', fontsize=18)
ax2 = plt.subplot(132)
sns.barplot(x=data2017_csump['region'],y=data2017_csump['total_water_consumption'],palette=sns.color_palette("Blues_r"))
ax2.set_title('2017 Reservoir Numbers', fontsize=18)
ax3 = plt.subplot(133)
sns.barplot(x=data2018_csump['region'],y=data2018_csump['total_water_consumption'],palette=sns.color_palette("Blues_r"))
ax3.set_title('2018 Reservoir Numbers', fontsize=18)
sns.despine()
```

```python
data2018_mostNumReservoir = (data2018.sort_values(by=['reservoir_numbers'],ascending=False)).head()
data2018_mostStorageReservoir = (data2018.sort_values(by=['reservoir_storage'],ascending=False)).head()
```

```python
plt.figure(figsize=(10,4), dpi=80)
plt.figure(1)
ax1 = plt.subplot(121)
sns.barplot(x=data2018_mostNumReservoir['region'],y=data2018_mostNumReservoir['reservoir_numbers'],palette=sns.color_palette("Blues_r"))
ax1.set_title('2018 Reservoir Numbers', fontsize=16)
ax1.set_ylabel('mumber')
ax2 = plt.subplot(122)
sns.barplot(x=data2018_mostStorageReservoir['region'],y=data2018_mostStorageReservoir['reservoir_storage'],palette=sns.color_palette("Blues_r"))
ax2.set_title('2018 Reservoir Storage', fontsize=16)
ax2.set_ylabel(' storage 10^8 m^3')
sns.despine()
```
